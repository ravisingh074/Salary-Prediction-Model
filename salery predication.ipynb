```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from word2number import w2n
```

```python
df = pd.read_excel(r"C:\Users\HP\Downloads\salary_sheet.xlsx")
```

```python
df
```

```python
df.experience = df.experience.fillna('zero')
df['experience']
```

```python
 df.experience = df.experience.apply(w2n.word_to_num)
df
```

```python
import math
```

```python
meadian_score = math.floor(df['test_score (out of 10)'].median())
```

```python
df['test_score (out of 10)'] = df['test_score (out of 10)'].fillna(meadian_score)
```

```python
df['test_score (out of 10)']
```

```python
reg =  linear_model.LinearRegression()
reg.fit(df[['experience','test_score (out of 10)','interview_score(out of 10)']],df['salary'])
```

```python
reg.predict([[0,10,10]])
```

```python
reg.predict([[10,5,5]])
```
